```python
#This is the code for the statistical analysis of the data
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
# Load data
df = pd.read_csv('/content/drive/My Drive/drugdata.csv')
df = df.iloc[:, :6].dropna()
```

```python
# Get statistical summary for each column
summary = df.describe()
print(summary)
```

```python
from google.colab import files
from PIL import Image
import os
for column in df.columns:
    plt.figure(figsize=(8.4, 4.2))
    plt.rcParams.update({'font.size': 18})
    sns.boxplot(x=df[column])  # Horizontal boxplot
    #plt.title(f'Boxplot of {column}')
    c = ''
    if column=='mm':
      c="Molar mass (g/mol)"
    elif column=='mp':
      c="Melting point (k)"
    elif column=='t':
      c="Temperature (k)"
    elif column=='p':
      c="Pressure (MPa)"
    elif column=='c':
      c="Solvent density(Kg/m3)"
    elif column=='y6':
      c="times6"
    else:
      c="Solubility"
    if c=="times6":
      plt.xlabel(r'Solubility ($\times 10^{-6}$)',fontsize=18)
    elif c=="Solvent density(Kg/m3)":
      plt.xlabel(r'Solvent  density (Kg/m$^{3}$)', fontsize=18)
    else:
      plt.xlabel(c, fontsize=18) # Set the x-axis label to the column name
    plt.tight_layout()  # Adjust layout to fit all labels
    temp_filename = c[:5]+'box.tiff'
    plt.savefig(temp_filename, dpi=600)

    img = Image.open(temp_filename)

    output_filename = c[:5]+'box_lzw.tiff'
    img.save(output_filename, compression='tiff_lzw')

    os.remove(temp_filename)

    from google.colab import files
    files.download(output_filename)
    plt.show()
```

```python
df_first_5 = df.iloc[:, :6]
# Dictionary mapping old column names to new ones
column_mapping = {
    'mm': 'Molar mass',
    'mp': 'Melting point',
    't': 'Temperature',
    'p': 'Pressure',
    'c': 'Solvent density',
    'y6': 'Solubility'
}
# Rename the columns using the mapping dictionary
plt.figure(figsize=(17.5, 17.5))
df_first_5.rename(columns=column_mapping, inplace=True)
plt.rcParams.update({'font.size': 18})
# Create the pairplot
sns.pairplot(df_first_5)
temp_filename = 'temp_pairplot.tiff'
plt.savefig(temp_filename, dpi=600)

img = Image.open(temp_filename)

output_filename = 'pairplot.tiff'
img.save(output_filename, compression='tiff_lzw')

os.remove(temp_filename)

from google.colab import files
files.download(output_filename)
plt.show()

plt.show()
```
